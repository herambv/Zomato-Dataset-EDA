```python
#Giving the average aggregate rating according to every country code

avgs={}
for i in data["Country Code"].unique():
    avg=round(np.mean(data.loc[data["Country Code"]==i]["Aggregate rating"]),4)
    avgs[i]=avg
avgs
```

```python
# Getting the country with the highest aggregate rating

avgs={}
for i in data["Country Code"].unique():
    ratings=data.loc[data["Country Code"]==i]["Aggregate rating"]
    avg=np.mean(ratings)
    avgs[i]=avg
key=[key for key,val in avgs.items() if val==max(avgs.values())]
print("The Country with code {} has the highest average rating with {}".format(key[0],round(max(avgs.values()),2)))
```

```python
# Making a function for getting the city with highest average rating from the country code

def max_average_ratings(country_code):
    cities=data.loc[data["Country Code"]==country_code]["City"].unique()
    avgs={}
    for i in cities:
        ratings=data.loc[data["City"]==i]["Aggregate rating"]
        avg=np.mean(ratings)
        avgs[i]=avg
    key=[key for key,val in avgs.items() if val==max(avgs.values())]
    print("The {} from Country {} has the highest average rating with {} ".format(key[0],country_code,round(max(avgs.values()),2)))
```

```python
# Fetching the city with highest aggregate rating from country code 162

max_average_ratings(162)
```

```python
# Fetching the cities with highest aggregate rating from every country code in the dataset

for i in data["Country Code"].unique():
    max_average_ratings(i)
    print()
```

```python
#Price range distribution (which shows the number of times that price range occured in the dataset)

plt.figure(figsize=(6,6))
plt.title("Price range distribution")
plt.pie(data["Price range"].value_counts(),labels=sorted(data["Price range"].unique()),autopct="%.2f")
print()
```

```python
# Average rating for every price range based on the countries

def price_range_country(country_code):    # function created to take country code as the input
    ranges=sorted(data["Price range"].unique())
    avgs={}
    for i in ranges:
        ratings=data.loc[(data["Price range"]==i) & (data["Country Code"]==country_code)]["Aggregate rating"]
        avg=round(np.mean(ratings),2)
        avgs[i]=avg
    print("For Country with code {} the average ratings for every price range is:".format(country_code))
    print(avgs)
price_range_country(1)
```

```python
#Aggregate rating according to every price range based on each city , given the country code

def price_range_city(country_code):
    ranges=sorted(data["Price range"].unique())
    cities=data.loc[data["Country Code"]==country_code]["City"].unique()
    avgs={}
    for city in cities:
        print("Average rating based on price range for {}:".format(city))
        for i in ranges:
            ratings=data.loc[(data["Price range"]==i) & (data["City"]==city)]["Aggregate rating"]
            avg=round(np.mean(ratings),2)
            avgs[i]=avg
        print(avgs)
        print()
price_range_city(162)
```

```python
#Average Cost For two according to every country

countries=[i for i in data["Country Code"].unique()]
for country in countries:
    print("Average Cost for two in coutry {} is: ".format(country))
    print(np.mean(data.loc[data["Country Code"]==country]["Average Cost for two"]),data.loc[data["Country Code"]==country]["Currency"].unique()[0])
    print()
```

```python
#Average Cost For two according to every city given the Country Code

def acft(country_code):
    cities=data.loc[data["Country Code"]==country_code]["City"].unique()
    for city in cities:
        print("Average Cost for two in {} is: ".format(city))
        print(round(np.mean(data.loc[data["City"]==city]["Average Cost for two"]),0),data.loc[data["City"]==city]["Currency"].unique()[0])
        print()
acft(162)
```

```python
#pie chart of Distribution of Cuisines for any country given the country code as parameter

def cuisine_distribution(code):
    plt.pie(data.loc[data["Country Code"]==code]["Cuisines"].value_counts()[:10],labels=data.loc[data["Country Code"]==code]["Cuisines"].value_counts().index.tolist()[:10])
    plt.title("Distribution of Cuisines in Country {}".format(code),fontsize=15)
    print()
```
